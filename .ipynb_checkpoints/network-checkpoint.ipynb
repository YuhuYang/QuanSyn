```python
###complex network approached in Quantitative Linguistics###
from conllu import parse_incr
from collections import Counter

sentences = open(r'D:\database\sud-treebanks-v2.12\SUD_Chinese-PUD\zh_pud-sud-test.conllu',"r",encoding="utf-8")


#词类 词形 词元 
def conllu2edges(data,direction=None,weight=None):
    edges = []
    for sent in parse_incr(data):
        for word in sent:
            if word['deprel'] not in ['punct','root']:
                gov = (sent[word['head']-1]['form'],sent[word['head']-1]['upos'])
                dep = (word['form'],word['upos'])
                edges.append((dep,gov))
    return edges
Counter(conllu2edges(sentences))
```
